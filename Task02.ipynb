```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv("C:/Users/admin/Downloads/titanic.csv")
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(titanic_df.head())
```

```python
titanic_df.info()
```

```python
num_rows = titanic_df.shape[0]
num_columns = titanic_df.shape[1]
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
# Summary statistics of numerical columns
print("\nSummary statistics of numerical columns:")
print(titanic_df.describe())
```

```python
# Check for missing values
print("\nMissing values in the dataset:")
print(titanic_df.isnull().sum())
```

```python
titanic_df.columns
```

```python
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df
```

```python
#removing the column Cabin
titanic_df= titanic_df.drop('Cabin', axis = 1)
titanic_df
```

```python
titanic_df['Embarked'].value_counts()
```

```python
#changing the data types for the columns Survived, PClass, Sex, Age and Embarked
titanic_df['Survived']=titanic_df['Survived'].astype('double')
titanic_df['Pclass']=titanic_df['Pclass'].astype('category')
titanic_df['Sex']=titanic_df['Sex'].astype('category')
titanic_df['Age']=titanic_df['Age'].astype('int')
titanic_df['Embarked']=titanic_df['Embarked'].astype('category')
titanic_df.info()
```

```python
# Countplot of survivors
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=titanic_df)
plt.title('Count of Survivors (0 = Not Survived, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

Interpretation for the above graph: This bar chart displays the count of survivors from the Titanic disaster, with 0 representing those who did not survive and 1 representing those who did survive.The large blue bar on the left represents the count of passengers who did not survive the disaster. This bar is significantly taller than the orange bar on the right, indicating that a much larger number of people perished compared to those who survived.The orange bar on the right represents the count of passengers who did survive the sinking of the Titanic.The orange bar on the right represents the count of passengers who did survive the sinking of the Titanic.

```python
# Pie chart of survival rate based on gender
plt.figure(figsize=(8, 6))
survival_gender = titanic_df.groupby('Sex')['Survived'].mean()
plt.pie(survival_gender, labels=survival_gender.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Survival Rate Based on Gender')
plt.axis('equal')
plt.show()
```

Interpretation for the above graph: This pie chart illustrates the survival rate based on gender from the given data. The larger blue slice, representing 79.7%, corresponds to the female survival rate, while the smaller red slice, representing 20.3%, corresponds to the male survival rate. The data suggests that the number of females survived are more than males.

```python
# Countplot of survivors based on passenger class
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Pclass', data=titanic_df)
plt.title('Survivors based on Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

Interpretation for the above graph: The graph has two sections, one for those who did not survive (0) and one for those who did survive (1).
For those who did not survive, the bars show that the highest number came from 3rd class, followed by 2nd class, and then 1st class passengers.However, for those who did survive, the majority were from 1st class, with significantly fewer survivors from 2nd class and an even smaller number from 3rd class.
The graph clearly illustrates the stark difference in survival rates based on passenger class, with 1st class passengers having a much higher likelihood of surviving compared to those in lower classes on the Titanic.

```python
# Distribution of passenger age
plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Interpretation for the above graph: In the graph,The x-axis represents the age groups, while the y-axis shows the count or frequency of passengers in each age group.
The distribution appears to be roughly unimodal, with the peak occurring around the age group of 20-30 years old. This suggests that a significant portion of the passengers fall within this age range.There is a noticeable skew towards the younger age groups, with a gradual decline in passenger counts as age increases beyond the peak. However, there is still a considerable number of passengers in the older age groups, although fewer than the younger groups.

```python
# Countplot of embarked port
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=titanic_df)
plt.title('Count of Passengers based on Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.show()
```

Interpretation for the above graph: This bar chart displays the count of passengers based on their embarked port from the Titanic dataset. The y-axis represents the passenger count, while the x-axis shows the different embarked ports, which are coded as 'C', 'Q', and 'S'.
The blue bar indicates that around 168 passengers embarked from the port coded as 'C', which likely stands for Cherbourg, France.
The orange bar represents a smaller number of passengers, around 77, who embarked from the port coded as 'Q', which could refer to Queenstown (now Cobh), Ireland.
The green bar, being the tallest, shows that the majority of passengers, approximately 644, embarked from the port coded as 'S', which most likely corresponds to Southampton, England.This graph provides insights into the distribution of passengers based on their embarkation points for the ill-fated Titanic voyage, with Southampton contributing the highest number of passengers.

```python
#count plot for Parch column
plt.figure(figsize=(8, 6))
sns.countplot(x='Parch', data=titanic_df, palette = 'viridis')
plt.title('Distribution of Parch')
plt.xlabel('Number of Parents/Children Aboard')
plt.ylabel('Count')
plt.show()
```

Interpretation for the above graph: This graph appears to represent the distribution of the number of parents/children aboard the Titanic. The x-axis shows the number of parents/children, ranging from 0 to 6, while the y-axis displays the count or frequency.
The most notable feature is the tall purple bar at 0, indicating that a large number of passengers, around 675, had no parents or children with them on the voyage.The shorter blue bar at 1 suggests that a smaller group, around 125 passengers, had either one parent or one child aboard.The teal bar at 2 is even smaller, implying that fewer passengers, likely around 50, had two parents/children accompanying them.The remaining bars at 3, 4, 5, and 6 parents/children aboard are extremely small, indicating that very few passengers traveled with that many family members.
This distribution suggests that the majority of passengers on the Titanic were either traveling alone or with at most one or two parents/children, while larger family groups were less common.

```python
# Boxplot of passenger class and age
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=titanic_df)
plt.title('Passenger Class and Age Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

Interpretation for the above graph: This graph appears to display the age distribution of passengers across different passenger classes on the Titanic. The x-axis represents the passenger class, with 1 being first class, 2 being second class, and 3 being third class. The y-axis shows the age range.
For first class passengers (1), the box plot shows that most passengers were relatively young, with the median age around 40 years old and a few outliers at higher ages.
Second class passengers (2) had a wider age distribution, with the median age slightly lower than first class, around 30 years old. There were also some younger and older outliers.Third class passengers (3) had the lowest median age, around 25 years old. However, the age range was quite wide, with several outliers at older ages, even extending beyond 80 years old.
This graph suggests that higher passenger classes on the Titanic tended to have older passengers on average, while lower classes had a wider age range but skewed towards younger ages overall.

```python
# Violin plot of fare distribution based on passenger class
plt.figure(figsize=(10, 6))
sns.violinplot(x='Pclass', y='Fare', data=titanic_df)
plt.title('Fare Distribution Based on Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

Interpretation for the above graph: A few key observations from the graph,The majority of passengers were in the 3rd class, as indicated by the tall peak on the left side of the graph.The fares for 3rd class passengers were relatively low and clustered around a single value, suggesting a standard or fixed fare for this class.The fares for 2nd class passengers, represented by the orange shape, had a wider range and were generally higher than those for 3rd class.The fares for 1st class passengers, indicated by the green shape on the right, had the highest variability, with some passengers paying significantly more than others in the same class.
This fare distribution likely reflects the socioeconomic status and amenities associated with each passenger class during that era of ocean travel. The graph provides insights into the economic diversity among the passengers on the Titanic, with the wealthier individuals occupying the higher classes and paying higher fares for more luxurious accommodations.

```python
# Pairplot of numerical variables
plt.figure(figsize=(10, 8))
sns.pairplot(titanic_df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']], hue='Survived')
plt.show()
```

Interpretation for the above graph: These are the set of plots appears to be visualizing different features and their relationships with survival in the Titanic dataset. The top left plot shows the distribution of passenger ages, with the orange and blue dots representing those who did not survive and survive, respectively. The age distribution seems to be slightly right-skewed.The top middle plot displays the SibSp feature, which represents the number of siblings or spouses aboard. The top right plot shows the Parch feature, which denotes the number of parents or children aboard. Both SibSp and Parch exhibit a large number of zeros, indicating many passengers were traveling alone.
The bottom left plot reveals that most passengers did not survive the tragedy, as indicated by the higher density of orange dots at 0 on the Survived axis. The bottom middle plot showcases the fare distribution, with a significant peak near 0, suggesting many passengers had low-cost tickets.Overall, these plots provide insights into the characteristics of the Titanic passengers and allow for exploring potential relationships between features like age, family size, fare, and survival outcomes.
